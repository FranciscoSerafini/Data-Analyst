operaciones con columnas
cargamos un archivo csv

```python

df_iris=pd.read_csv("data/iris.tab",sep="\t",skiprows=(1,2))
df_iris.head()
```

```python
df_iris.columns=["sepal_length", "sepal_width", "petal_length","petal_width", "iris"]
```

```python
df_iris.head()
df_iris.info()
```

multiplicamos una columna por un numero

```python

df_iris.sepal_length * 2
```

```python
df_iris.head()
```

sumamos y restamos columnas --> columna + columna
                                columna - columna

```python
df_iris.sepal_length - df_iris.sepal_width
```

```python
#agregamos esta columna nueva al data frame
df_iris["sp-sw"] = df_iris.sepal_length - df_iris.sepal_width
df_iris.head()
#en este caso estariamos agregando una columna nueva con la resta de las columnas 
```

```python
# T A R E A
#multiplicacion de columnas y division
#multiplicacion ----> columna * columna
#division -----> columna // columna
df_iris["sup-pet"]=df_iris.petal_length * df_iris.petal_width
df_iris.head()
```

```python
# T A R E A
#superficie para sepalos
df_iris["sup-sep"] = df_iris.sepal_length * df_iris.sepal_width
df_iris.head()
```

```python
# T A R E A
#division entre petalos
df_iris["pet_l/pet_w"] = df_iris.petal_length // df_iris.petal_width
df_iris.head()
```

```python
# T A R E A
#elevar al cuadrado los valores de petal lenght
df_iris.petal_length ** 2
```
