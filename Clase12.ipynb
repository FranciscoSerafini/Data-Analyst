```python
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
df_iris=pd.read_csv("data/iris.tab", sep="\t", skiprows=(1,2))
df_iris.columns=["sepal_lenght", "sepal_width", "petal_length", " petal_width", "iris"]
df_iris.head()
```

```python
#para graficar utilizaremos el kind = nos indica la forma del grafico
df_iris.plot(kind="scatter", x="sepal_lenght", y="sepal_width")
```

```python
#dentro de la documentacion podremos ver que hay mas atributos para nuestro grafico por ejemplo ponerle un titulo
df_iris.plot(kind="scatter", x='petal_length',y='petal_length', title="Grafico de petalos", figsize=[5,5])
```

```python
#creamos un data frame para visualizar lineas rectas y lineas con puntos
```

```python
x = np.arange(0,10,0.5) #creamos el siguiente rango de datos que vamos a utilizar
y= np.sin(x)

df=pd.DataFrame() #creamos el nuevo dataFrame
df["x"] = x #le damos los valores tanto a x como a y
df["y"] = y
df #esto nos va a dar una tabla de valores
```

```python
df.plot(kind="scatter", x="x", y="y") #le pasamos los distintos valores a x e y para crear el grafico
```

```python
df.plot(kind="line", x="x", y="y")
```

```python
#ordenamos los datos de sepal_lenght
#df_iris.sort_values("sepal_length").plot(kind="line", x="sepal_length", y="petal_length")
df_iris.sort_values('sepal_length').plot(kind='line',x='sepal_length',y='petal_length')
```

```python
#grafico de barras
df.plot(kind="bar", x="x", y="y")
```

```python
#trabajaremos con un group by que es para agrupar distintos tipos de datos
df_vuelos=pd.read_csv("data/flights.csv")
df_vuelos.head()
```

```python
df_vuelos.DAY.unique()
df_vuelos.MONTH.unique()
```

```python
df_vuelos.groupby("AIRLINE").count()["MONTH"]
```

```python
df_vuelos.groupby("AIRLINE").count()["DAY"].plot(kind="barh", figsize=[5,5])
```

```python
df_vuelos.groupby("AIRLINE").count()
```
